<h1>Introduction</h1>

<p>Data scientists can use various tools and techniques to explore, visualize, and manipulate data. One of the most common ways in which data scientists work with data is to use the Python language and some specific packages for data processing.</p>


<h2>What is NumPy?</h2>

<p>NumPy is a Python library that provides functionality comparable to mathematical tools such as MATLAB and R. While NumPy significantly simplifies the user experience, it also offers comprehensive mathematical functions.</p>

<h2>What is Pandas?</h2>

<p>Pandas is an extremely popular Python library for data analysis and manipulation. Pandas is like a spreadsheet application for Python, providing easy-to-use functionality for data tables.</p>

<h2>Testing hypotheses</h2>

<p>Data exploration and analysis is typically an iterative process, in which the data scientist takes a sample of data and performs the following kinds of tasks to analyze it and test hypotheses:</p>

<ul>1. Clean data to handle errors, missing values, and other issues.</ul>
<ul>2. Apply statistical techniques to better understand the data and how the sample might be expected to represent the real-world population of data, allowing for random variation.</ul>
<ul>3. Visualize data to determine relationships between variables, and in the case of a machine learning project, identify features that are potentially predictive of the label.</ul>
<ul>4. Revise the hypothesis and repeat the process.</ul>

<h2>1. Exploring data arrays with NumPy</h2>

```python
data = [50,50,47,97,49,3,53,42,26,74,82,62,37,15,70,27,36,35,48,52,63,64]
print(data)
```

```python
import numpy as np
grades = np.array(data)
print(grades)
```

```python
print (type(data),'x 2:', data * 2)
print('---')
print (type(grades),'x 2:', grades * 2)
```

```python
#shape of the array

grades.shape
```

```python
grades[0]
```

```python
grades.mean()
```

```python
# Define an array of study hours
study_hours = [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,15.5,
               13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

# Create a 2D array (an array of arrays)
student_data = np.array([study_hours, grades])

# display the array
student_data
```

```python
# Show shape of 2D array

student_data.shape
```

```python
# Show the first element of the 2D array

student_data[0][0]
```

```python
# Get the mean value of each sub-array
avg_study = student_data[0].mean()
avg_grade = student_data[1].mean()

print('Average study hours: {:.2f}\nAverage grade: {:.2f}'.format(avg_study, avg_grade))
```

<h3>Loading a DataFrame from a file</h3>

```python
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv
df_students = pd.read_csv('grades.csv',delimiter=',',header='infer')
df_students.head()
```

<h3>Handling missing values</h3>

```python
df_students.isnull()
```

```python
#calculate the total no of missing values

df_students.isnull().sum()
```

```python
df_students[df_students.isnull().any(axis=1)]
```

```python
#replace the missing value with the mean study hours

df_students.StudyHours = df_students.StudyHours.fillna(df_students.StudyHours.mean())
df_students
```

```python
#drop rows or columns that contain null values by using the dropna method

df_students=  df_students.dropna(axis=0, how='any')
df_students
```

<h3>Explore data in the DataFrame</h3>

```python
# Get the mean study hours using to column name as an index
mean_study = df_students['StudyHours'].mean()

# Get the mean grade using the column name as a property (just to make the point!)
mean_grade = df_students.Grade.mean()

# Print the mean study hours and mean grade
print('Average weekly study hours: {:.2f}\nAverage grade: {:.2f}'.format(mean_study, mean_grade))
```

```python
# Get students who studied for the mean or more hours

df_students[df_students.StudyHours > mean_study]
```

```python
# What was their mean grade?
df_students[df_students.StudyHours > mean_study].Grade.mean()
```

```python
passes  = pd.Series(df_students['Grade'] >= 60)
df_students = pd.concat([df_students, passes.rename("Pass")], axis=1)

df_students
```

```python
#how many students passed and failed

print(df_students.groupby(df_students.Pass).Name.count())
```

```python
#find the mean study time and grade for the groups of students who passed and failed the course

print(df_students.groupby(df_students.Pass)[['StudyHours', 'Grade']].mean())
```

```python
# Create a DataFrame with the data sorted by Grade (descending)
df_students = df_students.sort_values('Grade', ascending=False)

# Show the DataFrame
df_students
```
